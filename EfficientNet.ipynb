```python
import os
import torch
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torch.nn as nn
import torch.optim as optim
from torchvision import models
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from tqdm import tqdm
import copy
import time
import GPUtil
from torch.utils.data import random_split


# Устройство
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')


# Пути к директориям BIRD525
# train_dir = 'Birds_test/train'
# valid_dir = 'Birds_test/valid'
# test_dir = 'Birds_test/test'

# Пути к директориям BIRD200
train_dir = 'Birds_test_200/Train'
test_dir = 'Birds_test_200/Test/'

# Размер изображения
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 15
LEARNING_RATE = 1e-4


# Названия классов
name_classes = [name for name in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, name))]
num_classes = len(name_classes)
print(f"Классов: {num_classes}, Устройства: {device}")


# Аугментации и нормализация
train_transform = transforms.Compose([
    transforms.Resize(IMAGE_SIZE),
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_test_transform = transforms.Compose([
    transforms.Resize(IMAGE_SIZE),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])


# Загрузка датасетов
train_dataset = datasets.ImageFolder(train_dir, transform=train_transform)
# valid_dataset = datasets.ImageFolder(valid_dir, transform=val_test_transform)
test_dataset = datasets.ImageFolder(test_dir, transform=val_test_transform)

train_size = int(0.8 * len(train_dataset))
valid_size = len(train_dataset) - train_size
train_dataset, valid_dataset = random_split(train_dataset, [train_size, valid_size])

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)


print(f"Обучающая выборка (train): {len(train_dataset)} изображений")
print(f"Валидационная выборка (valid): {len(valid_dataset)} изображений")
print(f"Тестовая выборка (test): {len(test_dataset)} изображений")

# Модель
model = models.efficientnet_b0(pretrained=True)
model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
model = model.to(device)


# Потери и оптимизатор
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)



# Обучение и валидация
def train_and_validate(model, train_loader, valid_loader, criterion, optimizer, device, epochs, patience, save_path='best_model.pth'):
    train_loss_history = []
    valid_loss_history = []
    train_acc_history = []
    valid_acc_history = []

    best_acc = 0.0
    best_model_wts = copy.deepcopy(model.state_dict())
    patience_counter = 0

    for epoch in range(epochs):
        print(f"Epoch {epoch+1}/{epochs}")
        print("-" * 30)

        # Train phase
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs} [Train]'):
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        train_loss = running_loss / len(train_loader)
        train_acc = correct / total
        train_loss_history.append(train_loss)
        train_acc_history.append(train_acc)

        # Validation phase
        model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for images, labels in tqdm(valid_loader, desc=f'Epoch {epoch+1}/{epochs} [Valid]'):
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                running_loss += loss.item()
                _, preds = torch.max(outputs, 1)
                correct += (preds == labels).sum().item()
                total += labels.size(0)

        valid_loss = running_loss / len(valid_loader)
        valid_acc = correct / total
        valid_loss_history.append(valid_loss)
        valid_acc_history.append(valid_acc)

        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.4f}, Acc: {train_acc:.4f}, Val Loss: {valid_loss:.4f}, Acc: {valid_acc:.4f}")

        # Early stopping and checkpointing
        if valid_acc > best_acc:
            best_acc = valid_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            patience_counter = 0
            torch.save(model.state_dict(), save_path)
            print(f"✅ Saved new best model to {save_path}")
        else:
            patience_counter += 1
            print(f"⏸️  No improvement. Patience: {patience_counter}/{patience}")

        if patience_counter >= patience:
            print("⛔ Early stopping triggered.")
            break

        scheduler.step()

    model.load_state_dict(best_model_wts)
    return model, train_loss_history, valid_loss_history, train_acc_history, valid_acc_history



# Запуск обучения
model = model.to(device)
model, train_loss, val_loss, train_acc, val_acc = train_and_validate(
    model, train_loader, valid_loader, criterion, optimizer, device, epochs=EPOCHS, patience=5, save_path='best_model.pth'
)


# Графики
def plot_metrics(train_metric, val_metric, title, ylabel):
    plt.plot(train_metric, label='Train')
    plt.plot(val_metric, label='Validation')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid()
    plt.show()


plot_metrics(train_loss, val_loss, 'Loss per Epoch', 'Loss')
plot_metrics(train_acc, val_acc, 'Accuracy per Epoch', 'Accuracy')


# Тестирование
def test_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in tqdm(test_loader, desc='Testing'):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    acc = correct / total
    print(f"Test Accuracy: {acc:.4f}")

test_model(model, test_loader)
```

```python
import numpy as np

def show_predictions_by_class(model, data_loader, class_names, device, images_per_class=5):
    model.eval()
    class_to_images = {i: [] for i in range(len(class_names))}  # Класс -> список (изображение, истинный, предсказанный)

    with torch.no_grad():
        for images, labels in data_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            _, preds = torch.max(outputs, 1)

            for img, true_label, pred_label in zip(images, labels, preds):
                if len(class_to_images[true_label.item()]) < images_per_class:
                    class_to_images[true_label.item()].append((img.cpu(), true_label.item(), pred_label.item()))

            # Если собраны все нужные изображения — выход
            if all(len(imgs) >= images_per_class for imgs in class_to_images.values()):
                break

    # Визуализация
    for class_idx, images in class_to_images.items():
        plt.figure(figsize=(15, 3))
        for i, (img, true_label, pred_label) in enumerate(images):
            img = img.permute(1, 2, 0).numpy()
            img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
            img = np.clip(img, 0, 1)

            plt.subplot(1, images_per_class, i + 1)
            plt.imshow(img)
            plt.title(f"True: {class_names[true_label]}\nPred: {class_names[pred_label]}", fontsize=9)
            plt.axis('off')

        plt.suptitle(f'Class: {class_names[class_idx]}', fontsize=14)
        plt.tight_layout()
        plt.show()


show_predictions_by_class(model, test_loader, train_dataset.dataset.classes, device, images_per_class=5)
```

```python
from PIL import Image
import torch
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

def predict_image(image_path, model, class_names, device):
    model.eval()

    # Преобразования (должны совпадать с валидационными)
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ])

    # Загрузка и преобразование изображения
    image = Image.open(image_path).convert('RGB')
    input_tensor = transform(image).unsqueeze(0).to(device)

    # Предсказание
    with torch.no_grad():
        outputs = model(input_tensor)
        _, predicted = torch.max(outputs, 1)
        predicted_class = class_names[predicted.item()]

    # Визуализация
    img_vis = np.array(image.resize((224, 224))) / 255.0
    plt.imshow(img_vis)
    plt.title(f"Predicted: {predicted_class}")
    plt.axis('off')
    plt.show()

    return predicted_class
```

```python
import torch
import cv2
import numpy as np
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

def show_gradcam(image_path, model, class_names, device):
    model.eval()
    
    # Hook for gradients
    gradients = None
    def backward_hook(module, grad_input, grad_output):
        nonlocal gradients
        gradients = grad_output[0].detach()
    
    # Register hook on the last conv layer
    target_layer = None
    # Find the last convolutional layer in EfficientNet
    for module in model.features.modules():
        if isinstance(module, torch.nn.Conv2d):
            target_layer = module
    if target_layer is None:
        raise ValueError("No convolutional layer found in model.features")
    
    handle = target_layer.register_full_backward_hook(backward_hook)

    # Preprocess image
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                            [0.229, 0.224, 0.225])
    ])

    image = Image.open(image_path).convert('RGB')
    input_tensor = transform(image).unsqueeze(0).to(device)
    input_tensor.requires_grad = True

    # Forward pass
    output = model(input_tensor)
    pred_class = output.argmax(dim=1).item()
    class_name = class_names[pred_class]

    # Backward pass
    model.zero_grad()
    output[0, pred_class].backward()

    # Get activations (using forward hook)
    activations = None
    def forward_hook(module, input, output):
        nonlocal activations
        activations = output.detach()
    
    handle_forward = target_layer.register_forward_hook(forward_hook)
    _ = model(input_tensor)  # Run forward again to capture activations
    handle_forward.remove()
    
    if gradients is None or activations is None:
        handle.remove()
        raise RuntimeError("Failed to capture gradients or activations")

    # Grad-CAM calculation
    pooled_gradients = torch.mean(gradients, dim=[0, 2, 3])
    
    # Weight the activations by the gradients
    for i in range(activations.shape[1]):
        activations[:, i, :, :] *= pooled_gradients[i]
    
    heatmap = torch.mean(activations, dim=1).squeeze()
    heatmap = np.maximum(heatmap.cpu().numpy(), 0)
    heatmap = cv2.resize(heatmap, (224, 224))
    heatmap = (heatmap - heatmap.min()) / (heatmap.max() - heatmap.min())

    # Visualization
    img = np.array(image.resize((224, 224))) / 255.0
    heatmap = cv2.applyColorMap(np.uint8(255 * heatmap), cv2.COLORMAP_JET)
    heatmap = np.float32(heatmap) / 255
    superimposed_img = heatmap * 0.4 + img

    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(superimposed_img)
    plt.title(f"Grad-CAM: {class_name}")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
    
    # Remove hook
    handle.remove()

# Usage

show_gradcam(image_path, model, train_dataset.classes, device)
```

```python
import os

def rename_folders(directory):
    # Получаем список всех папок в указанной директории
    for folder_name in os.listdir(directory):
        folder_path = os.path.join(directory, folder_name)
        
        # Проверяем, что это папка и название соответствует шаблону "001.Name"
        if os.path.isdir(folder_path) and len(folder_name) > 4 and folder_name[3] == '.':
            # Формируем новое название (без первых 4 символов)
            new_name = folder_name[4:]
            new_path = os.path.join(directory, new_name)
            
            # Переименовываем папку
            os.rename(folder_path, new_path)
            print(f"Renamed: '{folder_name}' -> '{new_name}'")


# rename_folders('D:\\Projects\\birds 200_2\\birds_train')
```
